# Filtering, Renaming, Dropping
- data.column_name and data["column_name"] allow to select a single column from a DataFrame
- data[ ["column_1","column_2"] ] allows to subset several (in this case two) columns from the DataFrame
- data.iloc[] - allows subsetting rows and columns of a DataFrame using indicies
- data.loc[] - allows subsetting rows and columns both with indecies and labels
- data[] and data.loc[] support subsettin with logical (conditional) 
- data.rename()- changes the given columns or index label
- data.drop() - drop the given columns or index from the corresponding axis
- data['column_name'].unique() or data.column_name.unique() - returns all the unique values of the column
- data['column_name'].nunique() or data.column_name.nunique() - returns number of unique values in the column
- data['column_name'].value_counts() or data.column_name.value_counts() - returns the unique values in the column with their number of appearance in a descending order sorted by the count

# Merging
- DataFrame() - create a dataframe by suppliying data and column names as an input (can be done using dictionaries)
- concat() - append one dataframe below the other (or sideways), takes list of dataframes as an input
- join() - merge two dataframes based on index (can as well be done with merge(left_index=True))
- merge() - merge (inner, outer, left, right) two dataframe using a common (or prespecified) column or the index as the key

```python
df1 = pd.DataFrame({'make':['audi','bmw','volvo','renault'],'rating':[5,3,4,2]})
```

```python
df1
```

```python
# merge the dataset
# HINT: remember the SHIFT-tab shows the function details and explanation
df.merge(df1,how='left',on='make')
```

# Groupings



*   get_dummies() - produces a dataframe with dumy variables (1 if True, 0 otherwise)
*   groupby() - groups creates groups of a given varaible (as index) and populates all the columns with a given aggregating functino (mean in our case)
*   crosstab() - creates crosstabulations (same as pivot tables), aggregation function can be the mean, median, count etc.
*   pivot_table() - same crosstab, but faster

```python
df.groupby(["make"])['price'].mean()
```

```python
# two dimensional table which provides frequency of matching cases
pd.crosstab(index=df.make,columns=df.fuel_type).head()
```

```python
pd.pivot_table(data=df,index=["make"],columns="fuel_type",values="price",aggfunc="mean").fillna(0).head()
```

# Data Visualization



* plt.figure(figsize=(8,6)) - creates the figure istance allowing for customization of size (figsize in inches), figure color etc.

* plt.title() - provides a title for the figure
* plt.ylabel() - provides the label of Y axis
* plt.xlabel() - provides the label of X axis
* plt.show() - shows the figure
* plt.savefig() - saves the figure in a specified format and filename
* plt.hist() - draw a histogram
* plt.scatter() - draws a scatterplot
*  plt.subplot() - create subplots (plt.subplot(nrows=1,ncolumns=2,index=1) and 
* plt.subplot(nrows=1,ncolumns=2,index=2) will create 1 row and2 graphs)
* plt.boxplot() - the box extends from the lower to upper quartile values of the data, with a line at the median. The whiskers extend from the box to show the range of the data. Flier points are those past the end of the whiskers.














```python
plt.figure(figsize=(8,6))
plt.hist(df.price)
plt.title("Car Price Distribution")
plt.ylabel("Frequencies")
plt.xlabel("Price")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df.make, df.price, color='green')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df.horsepower,df.price, color='navy')
plt.title("Relationship between horsepower and price")
plt.xlabel("Horsepower")
plt.ylabel("Price")
plt.show()
```

```python
# color the graph based on the brand
plt.figure(figsize=(13,6))
sns.scatterplot(x="horsepower", y="price", hue="make", data=df)
plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.hist(df.price)
plt.title("Car Price Distribution")
plt.ylabel("Frequencies")
plt.xlabel("Price")

plt.subplot(1,2,2)
plt.scatter(df.horsepower,df.price)
plt.title("Relationship between horsepower and price")
plt.xlabel("Horsepower")
plt.ylabel("Price")
plt.show()
```
